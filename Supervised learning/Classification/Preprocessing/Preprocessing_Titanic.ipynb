Preprocessing titanic data set for Classification

```python
import pandas as pd #used for data analysis
import numpy as np  #used for performing mathematical operations on arrays

import matplotlib.pyplot as plt  #is used to plot visualization graphs
import seaborn as sns  #used for data visualization and data anlysis

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
file_path = '/content/titanic_original.csv'
df = pd.read_csv(file_path)
df_Tour = df.copy()
print(f'there are {df_Tour.shape[0]} rows , and {df_Tour.shape[1]} columns')
```

```python
df_Tour.head()
```

```python
print('number of rows: ', df_Tour.shape[0])
print('number of columns: ', df_Tour.shape[1])
print('features are: ' , df_Tour.columns.tolist())
print('\n')
```

```python
print('Duplicated values are: \n', df_Tour.duplicated().sum())
```

```python
print('Missing values are: \n', df_Tour.isnull().sum().sort_values(ascending=False))
```

```python
# Handling missing values
df_Tour = df_Tour.drop(columns=['body','cabin','boat','home.dest','name','ticket'])
```

```python
df_Tour.head()
```

```python
df_Tour['age'] = df_Tour['age'].fillna(df_Tour['age'].median())
```

```python
print('Missing values are: \n', df_Tour.isnull().sum().sort_values(ascending=False))
```

```python
df_Tour['embarked'].value_counts()
```

```python
df_Tour['embarked'] = df_Tour['embarked'].fillna('S')
```

```python
df_Tour['fare'] = df_Tour['fare'].fillna(method='ffill')
df_Tour['pclass'] = df_Tour['pclass'].fillna(method='ffill')
df_Tour['survived'] = df_Tour['survived'].fillna(method='ffill')
df_Tour['sex'] = df_Tour['sex'].fillna(method='ffill')
df_Tour['sibsp'] = df_Tour['sibsp'].fillna(method='ffill')
df_Tour['parch'] = df_Tour['parch'].fillna(method='ffill')
```

```python
print('Missing values are: \n', df_Tour.isnull().sum().sort_values(ascending=False))
```

```python
df_Tour.info()
```

```python
df_Tour.head()
```

```python
# Encoding Categorical Variables

df_Tour['embarked'] = df_Tour['embarked'].astype('category')
df_Tour['embarked'] = df_Tour['embarked'].cat.codes

df_Tour['sex'] = df_Tour['sex'].astype('category')
df_Tour['sex'] = df_Tour['sex'].cat.codes

df_Tour.head()
```

```python
# Saving the dataset as a CSV file
df_Tour.to_csv('clean-Titanic.csv', index=False)
```
